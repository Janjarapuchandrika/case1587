```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
chandu=pd.read_csv('energy.csv.csv')
print(chandu)
srh=chandu[chandu['State']=='Andhra']
rcb=chandu[chandu['State']=='Karnataka']
fig,diag=plt.subplots(2)
diag[0].plot(srh['Date'],srh['Consumption'],color='red')
diag[1].plot(rcb['Date'],rcb['Consumption'],color='yellow')
srh=chandu[chandu['State']=='Andhra']
rcb=chandu[chandu['State']=='Karnataka']
fig,diag=plt.subplots(2)
diag[0].bar(srh['Date'],srh['Consumption'],color='red')
diag[1].bar(rcb['Date'],rcb['Consumption'],color='yellow')
cse4='Kerala'
cse5='Delhi'
tdp=[cse4,cse5]
srh=chandu[chandu['State']=='Andhra']['Consumption'].sum()
rcb=chandu[chandu['State']=='Karnataka']['Consumption'].sum()
ysr=[srh,rcb]
fig,diag=plt.subplots(2)
diag[0].pie(ysr,labels=tdp)
```
